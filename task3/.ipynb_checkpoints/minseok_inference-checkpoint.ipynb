```python
# image_path = '/hub_data2/drone_2021sample/super_resolution_annotation_img/그린꿈나래반_2.png'
# image_path = '/hub_data2/drone_2021sample/super_resolution_annotation_img/306호수업준비실_1.png'
# image_path = '/hub_data2/drone_2021sample/super_resolution_annotation_img/306호수업준비실_2.png'
# image_path = '/hub_data2/drone_2021sample/super_resolution_annotation_img/자료준비실_2.png'
# image_path ='/hub_data2/drone_2021sample/super_resolution_annotation_img/제1교시상담실_2.png'
image_path ='/hub_data2/drone_2022sample/flight02/frame0175.png'
```

```python
with open(image_path, 'rb') as f:
    data = f.read()
encoded_img = np.fromstring(data, dtype = np.uint8)
image = cv2.imdecode(encoded_img, cv2.IMREAD_COLOR) ## (H,W,3) BGR
image = image[:, :, ::-1] ## # BGR to RGB
```

```python
## In Task 3
task3 = Task3(**vars(opt))

image = np.ascontiguousarray(image)
bboxes, polys, score_text = inference(task3.craft, image, task3.text_threshold,\
                                    task3.link_threshold, task3.low_text, task3.canvas_size, task3.mag_ratio, False, None)
image_show = image.copy()
crop_image_list = []
inter_cropped_img = []
count = 0
for bbox in bboxes:
    x_max = int(np.max(bbox[:,0]))
    x_min = int(np.min(bbox[:,0]))

    y_max = int(np.max(bbox[:,1]))
    y_min = int(np.min(bbox[:,1]))

    ## crop image processing ##
    try:
        cropped_image = image[y_min:y_max, x_min:x_max] ## [H,W,3]
        # cropped_image = np.flip(cropped_image)
        cropped_img = Image.fromarray(cropped_image)
        # cropped_img = cropped_img.convert('L') ## no rgb
        cropped_img = cropped_img.convert('RGB') ## rgb information
        
        ########################################################################
        inter_cropped_img.append(cropped_img.copy()) ## 
        # display(cropped_img.copy())
        ########################################################################
        
        cropped_img = task3.transform(cropped_img).unsqueeze(0) ## [1, 1, H, W] 
        crop_image_list.append(cropped_img)
        count +=1
        cv2.rectangle(image_show,(x_min, y_min),(x_max, y_max), (0, 255 , 0), 2)
        cv2.putText(image_show, str(count) , (x_min,y_min-5), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
        
    except:
        continue

```

```python
for plot_img in inter_cropped_img:
    plt.imshow(plot_img)
    plt.show()
```

```python
if len(crop_image_list)!=0:
    image_tensors = torch.cat(crop_image_list)
    batch_size = image_tensors.size(0)
    crop_image_input = image_tensors.cuda()
    
    length_for_pred = torch.IntTensor([task3.batch_max_length] * batch_size).cuda()
    text_for_pred = torch.LongTensor(batch_size,task3.batch_max_length + 1).fill_(0).cuda()

    ## Recognition
    preds = task3.wiw(crop_image_input, text_for_pred, is_train=False)

    # select max probabilty (greedy decoding) then decode index to character
    _, preds_index = preds.max(2)
    preds_txt = task3.converter.decode(preds_index, length_for_pred)
else:
    print("No text!")
```

```python
for i, text in enumerate(preds_txt):
    print(i+1, text.split('[s]')[0])
```
